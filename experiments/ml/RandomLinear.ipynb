```python

#https://tutorials.technology/tutorials/19-how-to-do-a-regression-with-sklearn.html
#https://www.youtube.com/watch?v=JcI5Vnw0b2c&index=2&list=PLQVvvaa0QuDfKTOs3Keq_kaG2P55YRn5v
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# let's create a linear function with some error called f
def f(x):
    res = x* 25 + 3
    error = res * random.uniform(-0.11, 0.11) # you can play with the error to see how it affects the result
    return res + error

values = []
# now using f we are going to create 300 values.
for i in range(0, 300):
    x = random.uniform(1, 1000)
    y = f(x)
    values.append((x, y))
    


v = np.array(values)



```

```python
df = pd.DataFrame(data=v)

dfout = df.values
#plt.scatter(df[:,0],df[:,1],  color='black')
plt.scatter(dfout[:,0],dfout[:,1],  color='black')

#print(v)

plt.show()
```
